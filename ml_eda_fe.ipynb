```python
#separating independant and dependant features
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
#splitting data
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=12)
```

```python
X_train
```

```python
y_train
```

```python
#standardize the datasets(feature scaling)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)#to avoid data leakage to model we use transform only
```

```python
X_train
```

```python
X_test
```

```python
## model training
from sklearn.linear_model import LinearRegression
regr= LinearRegression()
```

```python
regr
```

```python
regr.fit(X_train,y_train)
```

```python
#print coefficients and intercept
print(regr.coef_)
```

```python
print(regr.intercept_)
```

```python
# prediction for test data
reg_pred=regr.predict(X_test)
```

```python
reg_pred
```

```python
# assumptions of linear regression
plt.scatter(y_test,reg_pred)
plt.xlabel('test truth data')
plt.ylabel('pred data')
```

```python
#residuals
residuals=y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
##scatter plot with predictions and residuals(uniform distribution)
plt.scatter(reg_pred,residuals)
```

```python
#(no shape  means the model is good...)this is called uniform distributin
```

```python
##performance metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
#r square and adjstd r square
from sklearn.metrics import r2_score
score1=r2_score(y_test,reg_pred)
score1
```

```python
#adjusted r square
1-(1-score1)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
